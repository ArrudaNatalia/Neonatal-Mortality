```python
%cd /home/narruda/Merge_Neonatal/Merge_sample_90/Dataset_models/Dataset_without_ignored/By_UF
df_total = {}
for filename in glob.glob('*.csv'):
    df_total[filename[:-4]] = pd.read_csv(filename, sep=",", encoding = "latin-1",low_memory=False)   
```

create categorical variable - n_nu_peso

* 1: baixo peso (crianças com menos de 2500 g)
* 2: peso insuficiente (2500 g a 2999 g)
* 3: peso adequado (3000 g a 3999 g)
* 4: excesso de peso (4000g ou mais)

```python
for key, values in df_total.items():
    peso_bins = [0,2500, 3000, 4000,8000]
    peso_labels = ["1", "2", "3", "4"]
    df_total[key]['n_ct_peso'] = pd.cut(df_total[key]['n_nu_peso'], peso_bins, labels=peso_labels, 
                                        right=True, include_lowest=True)
    print(key, df_total[key]['n_ct_peso'].value_counts())
```
